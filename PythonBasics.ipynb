```python
# Write a function to find the square of a number.
def square(number):
    return number**2

square(5)    





# Create a function that checks whether a given number is even or odd.

def evenOrOdd(num):
    if num%2 == 0:
        return "Number is even"
    else:
        return "Number is odd"


evenOrOdd(2)        
        

# Write a function to return the largest of three numbers.

def largestOfThreeNumber(a,b,c):
    if a> b and a>c:
        return f"{a} is greater number"
    elif b>a and b>c:
        return f"{b} is greater number"
    else:
        return f"{c} is greater number"

largestOfThreeNumber(0,2,11)


# Create a function to calculate the sum of all numbers in a list.

def sumOfNumbers(numbers):
    sum = 0
    for i in numbers:
        sum += i
    return sum

sumOfNumbers([2,3,5])   

# Write a function to check if a given string is a palindrome.

def palindrome(string):
    rev = ""
    rev += string[::-1]
    if rev.casefold() == string.casefold():
        return f"{string} is palindrome of {rev}"
    else:
        return f"{string} is not palindrome of {rev}"
        

palindrome("Madam")

            
# Write a function that counts the number of words in a sentence.


def letterCounter(sentence):
    li = sentence.split()
    return len(li)

letterCounter("Hello maseera") 




# Create a function to find the second largest number in a list.

def secondLargestFromList(list):
    if len(list) <2:
        return f" list have less elements : {list}"
    else:
        rev = sorted(list , reverse = True)
        return f"The second largest element from {list} is {rev[1]}"
    

secondLargestFromList([2,13,11,15])


# Write a function to calculate the GCD (Greatest Common Divisor) of two numbers.


# Write a function to generate the first n numbers of the Fibonacci sequence.
# 0 1 1 2 3 5 8 
def fibonacci_sequence(num):
    if num <= 0:
        return []  # Return an empty list if num is 0 or negative
    elif num == 1:
        return [0]  # Return [0] for the first Fibonacci number
    elif num == 2:
        return [0, 1]  # Return [0, 1] for the first two Fibonacci numbers

    # Start the sequence with the first two Fibonacci numbers
    sequence = [0, 1]
    for i in range(2, num):  # Start from the 3rd Fibonacci number
        t = sequence[-1] + sequence[-2]  # Sum of the last two numbers
        sequence.append(t)

    return sequence
        

print(fibonacci_sequence(5))       
# Write a function that returns the factorial of a number using recursion
```
